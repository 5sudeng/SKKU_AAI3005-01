## 1-2. Python 조건문

### ✔︎ 조건문 기본

```python
# if문에 사용되는 조건식은 boolean 출력이며, 이것이 True일 때 해당 조건식에 해당된 코드가 실행됨.
var1 = 5
var2 = 10

print(var1 < var2)
```

```python
# if문에 비교 연산자를 통한 식을 기술, 조건에 부합할 때 print() 함수를 실행.
var1 = 5
var2 = 10

if var1 < var2:
  print("var2이 var1보다 큰 수입니다.")
```

```python
# if, else 문을 활용해 if~에 서술한 조건에 해당되지 않는 경우 실행될 코드 지정
var1 = 5
var2 = 10

if var1 > var2:
  print("var1이 var2보다 큰 수입니다.")
else: # var1 =< var2
  print("var2이 var1보다 큰 수입니다.")
```

```python
# if, elif, else 문을 활용해 세 가지 이상의 조건을 지정
var1 = 5
var2 = 5

if var1 > var2:
  print("var1이 var2보다 큰 수입니다.")
elif var1 == var2:
  print("var1과 var2는 같은 수입니다.")
else: # var1 < var2
  print("var2이 var1보다 큰 수입니다.")
```

### ✔︎ 조건문 응용
- 일반적으로 조건문에는 True, False와 같은 조건식이 들어오기 때문에 비교 연산자 등을 통해 조건식을 구성해주어야 함.
- 하지만 일반적으로 Python의 변수, 자료형 등은 비교 연산자 없이도 조건식에 True, False로 포함될 수 있음

```python
# 리스트를 활용한 조건문: 리스트에 하나 이상의 요소가 존재할 때 해당 리스트는 True를 나타냄.
temp = [1, 2, 3]
if temp: # True
  print("True")
```

```python
# 리스트를 활용한 조건문: 빈 리스트일 때, 해당 리스트는 False를 나타냄.
temp = []
if temp: # True
  print("True")
else: # False
  print("False")
```

```python
# 변수 값을 활용한 조건문: 변수 값이 0일 때, 해당 변수는 False를, 0이 아닐 때 해당 변수는 True를 나타냄.
temp_1 = 1
temp_0 = 0

if temp_1: # True
  print("True")
else:
  print("False")

if temp_0: # False
  print("True")
else:
  print("False")
```

## 1-5. Python 클래스와 인스턴스
- 클래스 : 특정한 기능을 수행하는 변수, 함수들의 집합틀 (붕어빵 틀)
- 인스턴스 : 클래스로부터 실제로 정의된 데이터 (붕어빵)

### ✔︎ Python 클래스 기본 예시

```python
class Calculator:
   def setdata(self, first, second):
       self.first = first
       self.second = second

   def add(self):
       result = self.first + self.second
       return result

   def mul(self):
       result = self.first * self.second
       return result

   def sub(self):
       result = self.first - self.second
       return result

   def div(self):
       result = self.first / self.second
       return result
```

```python
calculator = Calculator()
```

```python
calculator.setdata(10, 5)
```

```python
calculator.first, calculator.second
```

```python
print("두 숫자를 더한 결과는 : ", calculator.add())
print("두 숫자를 곱한 결과는 : ", calculator.mul())
print("두 숫자를 뺀   결과는 : ", calculator.sub())
print("두 숫자를 나눈 결과는 : ", calculator.div())
```

### ✔︎ Python 클래스 상속

```python
class Student:
    def __init__(self, name, id):
        self.name = name
        self.id = id

    def info(self):
        return f"이름: {self.name}, 학번: {self.id}"
```

```python
class FullTimeStudent(Student):
    def __init__(self, name, id, advisor):
        super().__init__(name, id)  # 부모 클래스의 생성자 호출
        self.advisor = advisor

    def info(self):
        return super().info() + f", 지도교수: {self.advisor}"
```

```python
class PartTimeStudent(Student):
    def __init__(self, name, id, company):
        super().__init__(name, id)  # 부모 클래스의 생성자 호출
        self.company = company

    def info(self):
        return super().info() + f", 회사: {self.company}"
```

```python
minhan = FullTimeStudent("Minhan Cho", "12345678", "Jinyoung Han")
print(minhan.info())

yumin = PartTimeStudent("Yumin Lee", "12345678", "Samsung")
print(yumin.info())
```

## 1-6. 연습문제 (사용자 정의 함수와 클래스)
"BankAccount" (계좌)라는 클래스를 작성하시오.

```python
#### Write your code ####
class BankAccount :
  def __init__(self, name, balance) :
    self.name = name
    self.balance = balance

  def deposit(self, amount) :
    self.balance += amount
    return self.balance

  def withdraw(self, amount) :
    self.balance -= amount
    if self.balance < 0 :
      print("insufficient funds")
      self.balance += amount
      return False
    else :
      return self.balance

  def display_balance(self) :
    return f"{self.name}'s balance: {self.balance}"

def transfer(sender, receiver, amount) :
  if sender.withdraw(amount) :
    receiver.deposit(amount)
```

```python
#########Do not edit here#############
john = BankAccount('John', 1000)
jane = BankAccount('Jane', 500)

transfer(john, jane, 200)

print(john.display_balance())  # 출력: 'John's balance: 800'
print(jane.display_balance())  # 출력: 'Jane's balance: 700'
```
